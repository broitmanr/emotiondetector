```python
import numpy as np

# Funciones de activación
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Datos de entrada y etiquetas
X = np.random.rand(48*48, 1)  # Datos de entrada (48*48 neuronas de entrada)
y = np.array([[0.5, 0.3, 0.1, 0.9]]).T  # Etiquetas (4 neuronas de salida)

# Inicialización de parámetros
np.random.seed(1)
input_neurons = 48*48
hidden_neurons1 = 128
hidden_neurons2 = 64
output_neurons = 4

# Pesos y sesgos
weights_input_hidden1 = np.random.uniform(-1, 1, (hidden_neurons1, input_neurons))
bias_hidden1 = np.random.uniform(-1, 1, (hidden_neurons1, 1))

weights_hidden1_hidden2 = np.random.uniform(-1, 1, (hidden_neurons2, hidden_neurons1))
bias_hidden2 = np.random.uniform(-1, 1, (hidden_neurons2, 1))

weights_hidden2_output = np.random.uniform(-1, 1, (output_neurons, hidden_neurons2))
bias_output = np.random.uniform(-1, 1, (output_neurons, 1))

# Hiperparámetros
learning_rate = 0.1
epochs = 1000

# Entrenamiento
for epoch in range(epochs):
    # Forward pass
    hidden1_input = np.dot(weights_input_hidden1, X) + bias_hidden1
    hidden1_output = sigmoid(hidden1_input)
    
    hidden2_input = np.dot(weights_hidden1_hidden2, hidden1_output) + bias_hidden2
    hidden2_output = sigmoid(hidden2_input)
    
    output_input = np.dot(weights_hidden2_output, hidden2_output) + bias_output
    output = sigmoid(output_input)
    
    # Cálculo de la pérdida
    loss = np.mean((output - y) ** 2)
    
    # Backward pass
    output_error = (output - y) * sigmoid_derivative(output)
    hidden2_error = np.dot(weights_hidden2_output.T, output_error) * sigmoid_derivative(hidden2_output)
    hidden1_error = np.dot(weights_hidden1_hidden2.T, hidden2_error) * sigmoid_derivative(hidden1_output)
    
    # Actualización de parámetros
    weights_hidden2_output -= learning_rate * np.dot(output_error, hidden2_output.T)
    bias_output -= learning_rate * np.sum(output_error, axis=1, keepdims=True)
    
    weights_hidden1_hidden2 -= learning_rate * np.dot(hidden2_error, hidden1_output.T)
    bias_hidden2 -= learning_rate * np.sum(hidden2_error, axis=1, keepdims=True)
    
    weights_input_hidden1 -= learning_rate * np.dot(hidden1_error, X.T)
    bias_hidden1 -= learning_rate * np.sum(hidden1_error, axis=1, keepdims=True)
    
    # Imprimir la pérdida en cada época
    if epoch % 100 == 0:
        print(f'Epoch {epoch}, Loss: {loss}')

# Predicción
print("\nPredicción final:")
print(output)
```
